## Fitting logistic regression model

### Fill NA

```python
data_all_na=data_all.copy()
```

Filling NA with **n/d** for categorical variables.

```python
categorical_vars=list(data_all_cnt[list(data_all_na.dtypes=='category')[1:]].alias)

for i in categorical_vars:
    data_all_na[i]=data_all_na[i].cat.add_categories('nd')
    data_all_na.loc[:,i].fillna('nd',inplace=True)
```

Filling NA with column **median** or **default min value (-999)** for remaining colums

```python
numerical_vars=data_all_cnt[list(data_all_na.dtypes!='category')[1:]].alias

#mins=pd.DataFrame(data_all_na.loc[:,numerical_vars].min()).min()
# min is -660 -> -999

for i in numerical_vars:
    data_all_na.loc[:,i].fillna(int(data_all_na.loc[:,i].astype(float).mean(axis = 0,skipna=True)),inplace=True)
    #data_all_na.loc[:,i].fillna(-999),inplace=True)    
    
```

### Encoding labels for categorial variables

Encoding categorical (string) labels to numbers to allow Logistic Regression

```python
le = LabelEncoder()
for i in categorical_vars:
    data_all_na[i]=le.fit_transform(data_all_na[i])
```

### Fit model

```python
Y=pd.DataFrame(data_all_na.loc[:,'isFraud'])
X=pd.DataFrame(data_all_na.iloc[:,2:data_all_na.shape[1]])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=100)
del Y, X, data_all_na
```

```python
%%time
#logreg = LogisticRegression(solver='saga', C=0.9 ,random_state=100)  
logreg = LogisticRegression(solver='lbfgs', random_state=100, max_iter=1000, C=0.7)  
# solver 'sag' and 'saga' quicker for large datasets
logreg.fit(X_train, np.ravel(Y_train.to_numpy()))
```

### Model diagnostics

```python
Y_scores=logreg.decision_function(X_test)
Y_predicted=logreg.predict(X_test)

fpr, tpr, _ = roc_curve(Y_test, Y_scores)
plt.plot(fpr, tpr, label='AUC - ROC score: %.4f' % roc_auc_score(Y_test, Y_scores))
plt.plot([0, 1], [0, 1], linestyle='--')
plt.title('ROC curve')
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.legend(loc='lower right')

print(classification_report(Y_test, Y_predicted))
```
